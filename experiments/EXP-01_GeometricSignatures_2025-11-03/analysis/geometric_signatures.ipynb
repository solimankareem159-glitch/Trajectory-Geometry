# Geometric Signatures in Embedding Space

This notebook analyzes how conversational moves create geometric signatures in embedding space.

```python
import google.generativeai as genai
import numpy as np
from scipy.spatial.distance import cosine, euclidean
```

```python
# Setup API Key
import os
GOOGLE_API_KEY = os.environ.get("GOOGLE_API_KEY", "YOUR_API_KEY_HERE")
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
def get_embedding(text, model="models/text-embedding-004"):
    """Get embedding for input text using the specified model."""
    result = genai.embed_content(model=model, content=text)
    return np.array(result['embedding'])
```

```python
def compute_distances(emb1, emb2):
    """Compute cosine and euclidean distances between two embeddings."""
    cos_dist = cosine(emb1, emb2)
    euc_dist = euclidean(emb1, emb2)
    return {
        "cosine_distance": cos_dist,
        "euclidean_distance": euc_dist
    }
```

```python
# Example Usage
text1 = "Hello, how are you?"
text2 = "Hi, how is it going?"

emb1 = get_embedding(text1)
emb2 = get_embedding(text2)

distances = compute_distances(emb1, emb2)
print(f"Distances between '{text1}' and '{text2}':")
print(distances)
```
